```python
import numpy as np
import math
import matplotlib.pyplot as plt
from numpy.linalg import solve
```

```python
def generate_points_sin(n):
    # Генерация n случайных точек для синуса
    x_values = np.linspace(-2, 2, n)
    y_values = np.sin(x_values)
    return x_values, y_values
```

```python
def generate_points_exp(n):
    # Генерация n случайных точек для экспоненты
    x_values = np.linspace(-2, 2, n)
    y_values = []
    for i in x_values:
        y_values.append(math.exp(i))
    #y_values = np.array(y_values)
    return x_values, y_values
```

```python
def generate_points_abs(n):
    # Генерация n случайных точек для модуля
    x_values = np.linspace(-2, 2, n)
    y_values = abs(x_values)
    return x_values, y_values
```

```python
def interpolate(x_values, y_values, n):
    
    # Подготовка матрицы для системы линейных уравнений
    A = np.vander(x_values, n, increasing=False)
    #print (A)

    # Решение системы линейных уравнений методом наименьших квадратов
    poly_coeffs = np.linalg.solve(A, y_values)
    #print (poly_coeffs)
    #print(np.poly1d(poly_coeffs))

    return np.poly1d(poly_coeffs)
```

```python
def plot_interpolation_sin(x, y, interpolated_function):
    # Построение графика
    x_range = np.linspace(-2.5, 2.5, 1000)
    plt.plot(x_range, np.sin(x_range), label='f(x) = sin(x)', color='blue')
    plt.scatter(x, y, color = 'red', label = 'Исходные точки')
    plt.plot(x_range, interpolated_function(x_range), label='Интерполяционный многочлен', linestyle='dashed', color='green')

    plt.title('Интерполяция функции f(x) = sin(x)')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.legend()
    plt.ylim(-2, 2)
    plt.grid(True)
    plt.show()
```

```python
def plot_interpolation_exp(x, y, interpolated_function):
    # Построение графика
    x_range = np.linspace(-2.5, 2.5, 1000)
    plt.plot(x_range, np.exp(x_range), label='f(x) = exp(x)', color='blue')
    plt.scatter(x, y, color='red', label='Исходные точки')
    plt.plot(x_range, interpolated_function(x_range), label='Интерполяционный многочлен', linestyle='dashed', color='green')
    plt.title('Интерполяция функции f(x) = exp(x)')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.legend()
    plt.ylim(-1, 10)
    plt.grid(True)
    plt.show()
```

```python
def plot_interpolation_abs(x, y, interpolated_function):
    # Построение графика
    x_range = np.linspace(-2.5, 2.5, 1000)
    plt.plot(x_range, np.abs(x_range), label='f(x) = abs(x)', color='blue')
    plt.scatter(x, y, color='red', label='Исходные точки')
    plt.plot(x_range, interpolated_function(x_range), label='Интерполяционный многочлен', linestyle='dashed', color='green')
    plt.title('Интерполяция функции f(x) = abs(x)')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.legend()
    plt.ylim(-3, 3)
    plt.grid(True)
    plt.show()
```

```python
n = 50
x_values, y_values = generate_points_sin(n)
interpolated_function = interpolate(x_values, y_values, n)

plot_interpolation_sin(x_values, y_values, interpolated_function)
```

```python
n = 50
x_values, y_values = generate_points_abs(n)
interpolated_function = interpolate(x_values, y_values, n)

plot_interpolation_abs(x_values, y_values, interpolated_function)
```

```python
n = 50
x_values, y_values = generate_points_exp(n)
interpolated_function = interpolate(x_values, y_values, n)

plot_interpolation_exp(x_values, y_values, interpolated_function)
```
